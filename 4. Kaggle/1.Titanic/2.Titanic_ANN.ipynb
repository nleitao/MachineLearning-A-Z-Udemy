```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np
```

```python
#explore the data
dataset_train=pd.read_csv('./input/train.csv')
dataset_train.head()
```

```python
#define the y_vector
y_train=dataset_train.iloc[:,1].values

#see the y_vector
y_train[:5]
```

```python
#remove the y_vector from the dataset
X_train=dataset_train.drop('Survived',axis=1)

#remove column "Name" from X_train matrix. I think it won't add any value to the model
X_train=X_train.drop('Name',axis=1)

#remove column "Ticket" from X_train matrix. I think the ticket number won't add any value to the model
X_train=X_train.drop('Ticket',axis=1)

#remove column "Cabin" from X_train matrix. Not sure if cabin number matter but I've removed it just to simplify the model at 
#this stage. Because there are some entries with 3 cabin numbers.
X_train=X_train.drop('Cabin',axis=1)

#remove column "PassengerId" from X_train matrix.
X_train=X_train.drop('PassengerId',axis=1)

X_train.head()
```

```python
#Ok, now I think I have to work on 2 categorical variables: "Sex" and "Embarked"

label_encoder_X=LabelEncoder()
X_train.iloc[:,1]=label_encoder_X.fit_transform(X_train.iloc[:,1])
#Male are labeled now as 1 and Female as 0

label_encoder_for_Embarked=LabelEncoder()
#Force convert column to str because there are some NaN
X_train.iloc[:,6] = label_encoder_for_Embarked.fit_transform(X_train.iloc[:,6].astype(str))
X_train.iloc[:,6]=label_encoder_for_Embarked.fit_transform(X_train.iloc[:,6])
#Embarked at S is now encoded as 2. C = 0. O outro enconded as 1 (acho eu)

#Since I don't know now what to do with missing data. I'll just delete column "Age" as well.
X_train=X_train.drop('Age',axis=1)


X_train.head()
```

```python
#just in case you want to see NaN values and where they are located
#df = X_train[X_train.isnull().any(axis=1)]
#df
```

```python
#como tenho diferentes variaveis com diferentes magnitudes preciso de as normalizar
sc_X=StandardScaler()
X_train = sc_X.fit_transform(X_train)
```

```python
X_train
```

```python
X_train.shape
```

```python
#Make the Artificial Neural Network

#Importing Keras libraries and required packages
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
#Initialising the ANN
classifier=Sequential()

#Adding the input layer and the first hidden layer
classifier.add(Dense(output_dim=6, init='uniform', activation='relu',input_dim=6))
#So precisamos de definir o input layer no primeiro layer, dps disso a ANN descobre automaticamente

#uma tip que deram para o output_dim e a media da soma dos inputs parameters (11) com os output parameters (1).
#neste caso (11+1) / 2 = 6

#the inits and activations and optimizers and losses are different if instead of a binary classification we had more categories
#Adding the second hidden layer
classifier.add(Dense(output_dim=6, init='uniform', activation='relu'))

#Adding the output layer
classifier.add(Dense(output_dim=1, init='uniform', activation='sigmoid'))

#Compiling the ANN 
classifier.compile(optimizer='adam', loss='binary_crossentropy',metrics=['accuracy'])
```

```python
#Fitting the ANN to the training set
classifier.fit(X_train,y_train,batch_size=10, nb_epoch=100)
```

```python
#agora que treinei o modelo vou usar a matrix X_train para o testar (sim, a mesma que usei para treinar...)
y_pred=classifier.predict(X_train)
y_pred[:5]

#devolve uma probabilidade acho eu...
```

```python
#para fazer a confusion matrix escolho um treshold onde digo que acima desse valor retorna true (neste caso, a pessoa sobreviveu)
y_pred=(y_pred>0.5)

y_pred[:5]
```

```python
from sklearn.metrics import confusion_matrix

#vou comparar os resultados do meu modelo com os resultados reais fazendo uma confusion matrix
cm=confusion_matrix(y_train,y_pred)

cm

#calcular a accuracy do modelo
# (498+230)/(498+51+112+230)
```
