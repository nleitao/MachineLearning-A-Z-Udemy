```python
import pandas as pd
import matplotlib.pyplot as plt
import scipy
import numpy as np
```

```python
dataset=pd.read_csv('./../0. DataSets/Market_Basket_Optimisation.csv',header=None)
```

```python
dataset.tail()

#Este dataset contem o numero total de transacoes numa semana num supermercado (numero total de transacoes 7500)
#Pelos vistos cada linha e uma transacao/compras de uma pessoa, ou seja uma transacao
#Por exemplo linha 1: uma pessoa foi ao supermercado e comprou Burguers, meatballs e eggs
```

```python
#Este apriori algoritmo expects a list of lists as input and not a dataframe
#Basicamente o gajo so fez 2 for loops para transformar o dataframe numa lista de listas

transactions=[]
for i in range(0,7501):
    transactions.append([str(dataset.values[i,j]) for j in range(0,20)])
```

```python
#Training Apriori on the dataset

#importar o ficheiro dado na aula com o codigo necessario para trainar datasets neste apriori algorythm
import sys
sys.path.insert(0, './../3. Code_Modules/')
from apyori import apriori

rules=apriori(transactions, min_support=0.003, min_confidence=0.2, min_lift=3, min_length=2)

#min_length=2 // Quero encontrar 2 productos com relacao do genero quando comprado X foi tambem comprado Y

#min_support= 0.003 // Suporte definido por numero de ocurrencias dividido pelo numero total de transacoes
#Como quero um produto que foi comprado pelo menos 3x por dia durante essa semana= 3x7/7500 = 0.003

#min_confidence // A percentagem que quero que esta regra seja verdade. Mas cuidado, pois pode haver artigos que sao smpr comprados independentemente da nossa regra, por exemplo leite e pao sao smpr comprados
#it's better to chose a lower confidence value and then delete the products that we think are not explained by this rule

#min_lift // 
```
